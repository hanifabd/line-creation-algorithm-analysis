```python
import sys
import time
import os
import psutil
from OpenGL import *
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
```

```python
window = 0
width, height = 800,600
```

```python
def ddaAlgo(x1, y1, x2, y2):
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    
    x, y = x1, y1
    glVertex2f(x, y)
    
    if dx > dy:
        Step = dx
    else:
        Step = dy
    
    xInc = dx/Step
    yInc = dy/Step
    
#     for k in range(2, dx):
#         x = x + xInc
#         y = y + yInc
#         glVertex2f(x, y)
        
    while True:
        x = x + xInc
        y = y + yInc
        xRound = round(x)
        yRound = round(y)
        glVertex2f(xRound, yRound)
        if(xRound == x2 and yRound == y2):
            break
```

```python
def lineDda():
    glClear(GL_COLOR_BUFFER_BIT|GL_DEPTH_BUFFER_BIT)
    glBegin(GL_LINES)
    glColor(1.0, 1.0, 0.0)

    ddaAlgo(2, 1, 400, 250)
        
    glEnd()
    glutSwapBuffers()
```

```python
def main():
    glutInit(sys.argv)
    glutInitDisplayMode(GLUT_RGBA|GLUT_DOUBLE|GLUT_ALPHA|GLUT_DEPTH)
    glutInitWindowSize(width, height)
    glutInitWindowPosition(0,0)
    glutCreateWindow(b'line Dda')
    glClearColor(0.0, 0.0, 0.0, 0.0)
    gluOrtho2D(0.0, 500.0, 0.0, 400.0)
    glutDisplayFunc(lineDda)
    glutIdleFunc(lineDda)
    glutMainLoop()
```

```python
main()
```
